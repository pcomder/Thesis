```python
"""
MODEL 1 – 6-Year Panel (2018-2023)
Same-year ESG composite → Stock Return
"""

# 1 ▸ Imports & style
import pandas as pd, numpy as np, statsmodels.api as sm, matplotlib.pyplot as plt
import seaborn as sns, warnings; from linearmodels import PanelOLS; from scipy import stats
warnings.filterwarnings("ignore"); sns.set_style("whitegrid")

# 2 ▸ File paths
PRICES = "/content/year_end_closing_prices.csv"
ESG    = "/content/ESG_Data.csv"
CTRL   = "/content/control_variables.csv"

# 3 ▸ Helpers
def safe_split(x):
    return x.split()[0] if isinstance(x, str) and x.split() else str(x)

def safe_numeric(x):
    if pd.isna(x):
        return np.nan
    if isinstance(x, str):
        x = "".join(c for c in x if c.isdigit() or c in ".-")
    try:
        return float(x)
    except (ValueError, TypeError):
        return np.nan

# 4 ▸ Load & preprocess (2018-2023)
def load_raw():
    return (pd.read_csv(PRICES),
            pd.read_csv(ESG),
            pd.read_csv(CTRL))

def preprocess(prices, esg, ctrl):
    esg["Company"]  = esg["Ticker"].apply(safe_split)
    ctrl["Company"] = ctrl["Ticker"].apply(safe_split)

    yrs = [y for y in ["2017","2018","2019","2020","2021",
                       "2022","2023"] if y in prices.columns]

    # Returns 2018-23
    for i in range(1, len(yrs)):
        prices[f"Return_{yrs[i]}"] = (prices[yrs[i]] - prices[yrs[i-1]]) / prices[yrs[i-1]]
    ret = pd.melt(
        prices[["Company"] + [f"Return_{y}" for y in yrs[1:]]],
        id_vars="Company", var_name="Year", value_name="Return"
    ).assign(Year=lambda d: d.Year.str.replace("Return_",""))

    # ESG 2018-23
    esg_long = pd.DataFrame()
    for y in yrs[1:]:
        col = f"ESG - {y}"
        if col in esg.columns:
            i  = esg.columns.get_loc(col)
            tmp = esg[["Company", col,
                        esg.columns[i+1], esg.columns[i+2], esg.columns[i+3]]]
            tmp.columns = ["Company","ESG","ENV","SOC","GOV"]
            for c in ["ESG","ENV","SOC","GOV"]:
                tmp[c] = tmp[c].apply(safe_numeric)
            tmp["Year"] = y
            esg_long = pd.concat([esg_long, tmp])

    # Controls 2018-23
    ctrl_long = pd.DataFrame()
    for y in yrs[1:]:
        cols, ren = ["Company"], {}
        for raw,new in [
            (f"Market Cap - {y}",   "Market_Cap"),
            (f"P/E - {y}",          "PE_Ratio"),
            (f"ROA - {y}",          "ROA"),
            (f"Debts to Assets - {y}","Debt_to_Assets")]:
            if raw in ctrl.columns:
                cols.append(raw); ren[raw] = new
        tmp = ctrl[cols].rename(columns=ren)
        for c in ["Market_Cap","PE_Ratio","ROA","Debt_to_Assets"]:
            tmp[c] = tmp.get(c, np.nan).apply(safe_numeric)
        tmp["Year"] = y
        ctrl_long = pd.concat([ctrl_long, tmp])

    panel = (ret.merge(esg_long,  on=["Company","Year"])
                .merge(ctrl_long, on=["Company","Year"]))
    panel["Year"] = panel["Year"].astype(int)

    panel["ln_MCap"] = np.log(
        pd.to_numeric(panel["Market_Cap"], errors="coerce")
          .replace({0:np.nan}).mask(panel["Market_Cap"]<=0)
    )

    # Industry & year dummies
    np.random.seed(42)
    inds = ["Tech","Health","Fin","Cons","Energy"]
    panel["Industry"] = panel["Company"].map(
        {c: inds[i % 5] for i,c in enumerate(panel["Company"].unique())})
    panel = pd.concat(
        [panel,
         pd.get_dummies(panel["Industry"], prefix="ind",  drop_first=True),
         pd.get_dummies(panel["Year"],     prefix="year", drop_first=True)], axis=1)

    panel = panel[(panel[["ESG","ENV","SOC","GOV"]] < 10).all(axis=1)]
    return panel.set_index(["Company","Year"])

# 5 ▸ Random-effects regression
def run_re(panel):
    core = ["ESG","PE_Ratio","Debt_to_Assets","ROA","ln_MCap"]
    y    = panel["Return"]
    X    = sm.add_constant(panel[core +
             [c for c in panel.columns if c.startswith(("ind_","year_"))]])
    res  = PanelOLS(y, X, entity_effects=False).fit(
               cov_type="clustered", cluster_entity=True)
    return res, ["ESG"]

# 6 ▸ Hausman (robust, no intercept)
def hausman(panel, dep, vars_):
    y, X = panel[dep], panel[vars_]
    fe   = PanelOLS(y, X, entity_effects=True ).fit(
               cov_type="clustered", cluster_entity=True)
    re   = PanelOLS(y, X, entity_effects=False).fit(
               cov_type="clustered", cluster_entity=True)
    diff = fe.params - re.params
    V    = fe.cov - re.cov
    chi2 = float(diff.T @ np.linalg.pinv(V) @ diff)
    pval = 1 - stats.chi2.cdf(chi2, len(vars_))
    return chi2, pval

# 7 ▸ Plots
def plot_panel(p):
    fig, ax = plt.subplots(2,2, figsize=(16,11))

    sns.scatterplot(data=p.reset_index(), x="ESG", y="Return",
                    hue="Industry", ax=ax[0,0])
    ax[0,0].set_title("ESG vs Return")

    sns.boxplot(data=p.reset_index(), x="Industry", y="Return", ax=ax[0,1])
    ax[0,1].set_title("Return by Industry")
    plt.setp(ax[0,1].get_xticklabels(), rotation=45)

    melt = p.reset_index().melt(
        id_vars=["Company","Year","Industry","Return"],
        value_vars=["ENV","SOC","GOV"],
        var_name="Comp", value_name="Score")
    sns.scatterplot(data=melt, x="Score", y="Return",
                    hue="Comp", ax=ax[1,0])
    ax[1,0].set_title("ESG Components vs Return")

    (p.reset_index()
       .groupby("Year")[["ESG","ENV","SOC","GOV"]]
       .mean().plot(marker="o", ax=ax[1,1]))
    ax[1,1].set_title("Average ESG & Components by Year")

    plt.tight_layout(); plt.show()

# 8 ▸ Main
if __name__ == "__main__":
    prc, esg, ctl = load_raw()
    pnl = preprocess(prc, esg, ctl)

    re_res, test = run_re(pnl)
    print("\n=== MODEL 1  (Random-Effects, 2018-23) ===\n")
    print(re_res)

    chi2, p = hausman(pnl, "Return", test)
    print(f"\nHausman χ² = {chi2:.2f}   p-value = {p:.3f}")

    plot_panel(pnl)
```

```python
"""
MODEL 3  ─  6-Year Panel (2018-2023)
Same-year ESG ➜ Systematic Risk (Beta proxy)
"""

# 1 ▸ Imports & style
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from linearmodels import PanelOLS
from scipy import stats

warnings.filterwarnings("ignore")
sns.set_style("whitegrid")

# 2 ▸ File paths (edit if needed)
PRICES = "/content/year_end_closing_prices.csv"
ESG    = "/content/ESG_Data.csv"
CTRL   = "/content/control_variables.csv"

# 3 ▸ Helper functions
def safe_split(x):
    """Keep first token of a ticker string."""
    return x.split()[0] if isinstance(x, str) and x.split() else str(x)

def safe_numeric(x):
    """Convert strings like '$1,234.5' to float, else NaN."""
    if pd.isna(x):
        return np.nan
    if isinstance(x, str):
        x = "".join(c for c in x if c.isdigit() or c in ".-")
    try:
        return float(x)
    except (ValueError, TypeError):
        return np.nan

# 4 ▸ Load & preprocess  (builds 2018-2023 panel and Beta)
def load_raw():
    return (pd.read_csv(PRICES),
            pd.read_csv(ESG),
            pd.read_csv(CTRL))

def preprocess(prices, esg, ctrl):
    esg ["Company"] = esg ["Ticker"].apply(safe_split)
    ctrl["Company"] = ctrl["Ticker"].apply(safe_split)

    yrs = [y for y in ["2017","2018","2019","2020","2021","2022","2023"]
           if y in prices.columns]

    # ▸ Returns 2018-23
    for i in range(1, len(yrs)):
        prices[f"Return_{yrs[i]}"] = (prices[yrs[i]] - prices[yrs[i-1]]) / prices[yrs[i-1]]
    ret = pd.melt(
        prices[["Company"]+[f"Return_{y}" for y in yrs[1:]]],
        id_vars="Company", var_name="Year", value_name="Return"
    ).assign(Year=lambda d: d.Year.str.replace("Return_",""))

    # ▸ ESG 2018-23
    esg_long = pd.DataFrame()
    for y in yrs[1:]:
        col = f"ESG - {y}"
        if col in esg.columns:
            tmp = esg[["Company", col]].rename(columns={col: "ESG"})
            tmp["ESG"]  = tmp["ESG"].apply(safe_numeric)
            tmp["Year"] = y
            esg_long = pd.concat([esg_long, tmp])

    # ▸ Controls 2018-23
    ctrl_long = pd.DataFrame()
    for y in yrs[1:]:
        cols, ren = ["Company"], {}
        for raw, new in [(f"Market Cap - {y}",   "Market_Cap"),
                         (f"P/E - {y}",          "PE_Ratio"),
                         (f"ROA - {y}",          "ROA"),
                         (f"Debts to Assets - {y}", "Debt_to_Assets")]:
            if raw in ctrl.columns:
                cols.append(raw)
                ren[raw] = new
        tmp = ctrl[cols].rename(columns=ren)
        for c in ["Market_Cap","PE_Ratio","ROA","Debt_to_Assets"]:
            tmp[c] = tmp.get(c, np.nan).apply(safe_numeric)
        tmp["Year"] = y
        ctrl_long = pd.concat([ctrl_long, tmp])

    # ▸ Merge to panel
    panel = (ret
             .merge(esg_long,  on=["Company","Year"])
             .merge(ctrl_long, on=["Company","Year"]))
    panel["Year"] = panel["Year"].astype(int)

    # ▸ ln(Market Cap)
    panel["ln_MCap"] = np.log(
        pd.to_numeric(panel["Market_Cap"], errors="coerce")
          .replace({0: np.nan}).mask(panel["Market_Cap"]<=0)
    )

    # ▸ Industry & year dummies
    np.random.seed(42)
    inds = ["Tech","Health","Fin","Cons","Energy"]
    panel["Industry"] = panel["Company"].map(
        {c: inds[i % 5] for i,c in enumerate(panel["Company"].unique())})
    panel = pd.concat(
        [panel,
         pd.get_dummies(panel["Industry"], prefix="ind",  drop_first=True),
         pd.get_dummies(panel["Year"],     prefix="year", drop_first=True)], axis=1)

    # ▸ Beta proxy
    panel["Beta"] = panel["Return"] / panel.groupby("Year")["Return"].transform("mean")

    return panel.set_index(["Company","Year"])

# 5 ▸ Random-effects regression
def run_re(panel):
    core = ["ESG","PE_Ratio","Debt_to_Assets","ROA","ln_MCap"]
    y = panel["Beta"]
    X = sm.add_constant(
        panel[core + [c for c in panel.columns if c.startswith(("ind_","year_"))]]
    )
    res = PanelOLS(y, X, entity_effects=False).fit(
        cov_type="clustered", cluster_entity=True
    )
    return res, ["ESG"]

# 6 ▸ Hausman test (robust, no intercept)
def hausman(panel, dep, vars_):
    y, X = panel[dep], panel[vars_]
    fe = PanelOLS(y, X, entity_effects=True ).fit(
        cov_type="clustered", cluster_entity=True)
    re = PanelOLS(y, X, entity_effects=False).fit(
        cov_type="clustered", cluster_entity=True)
    diff = fe.params - re.params
    V    = fe.cov - re.cov
    chi2 = float(diff.T @ np.linalg.pinv(V) @ diff)
    pval = 1 - stats.chi2.cdf(chi2, len(vars_))
    return chi2, pval

# 7 ▸ Diagnostic plots
def plot_panel(p):
    fig, ax = plt.subplots(2,2, figsize=(16,11))

    sns.scatterplot(data=p.reset_index(), x="ESG", y="Beta",
                    hue="Industry", ax=ax[0,0])
    ax[0,0].set_title("Beta vs ESG")

    sns.boxplot(data=p.reset_index(), x="Industry", y="Beta", ax=ax[0,1])
    ax[0,1].set_title("Beta by Industry")
    plt.setp(ax[0,1].get_xticklabels(), rotation=45)

    sns.scatterplot(data=p.reset_index(), x="ln_MCap", y="Beta",
                    hue="Industry", ax=ax[1,0])
    ax[1,0].set_title("Beta vs ln(MCap)")

    (p.reset_index()
       .groupby("Year")["Beta"]
       .mean().plot(marker="o", ax=ax[1,1]))
    ax[1,1].set_title("Average Beta by Year")

    plt.tight_layout()
    plt.show()

# 8 ▸ Run the whole pipeline
if __name__ == "__main__":
    prc, esg, ctl = load_raw()
    pnl = preprocess(prc, esg, ctl)

    re_res, test = run_re(pnl)
    print("\n=== MODEL 3  (Random-Effects, 2018-23) ===\n")
    print(re_res)

    chi2, p = hausman(pnl, "Beta", test)
    print(f"\nHausman χ² = {chi2:.2f}   p-value = {p:.3f}")

    plot_panel(pnl)
```
